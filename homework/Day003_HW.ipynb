### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
import os, sys

# 打开文件
dirs = os.listdir( './data' )

# 输出所有文件和文件夹
for file in dirs:
    print(file)
```

```python
#File I/O
# 讀檔案
fh = open("./data/64_72hr_CH.xml", "r", encoding = 'utf8')
xml = fh.read()
fh.close()

print(xml)
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
with open('./data/64_72hr_CH.xml',encoding='utf8') as fh:
    xml = fh.read()
    d = dict(xmltodict.parse(xml))
    locations = d['cwbopendata']['dataset']['locations']['location']
    print(len(locations))
    #print(d['cwbopendata']['dataset']['locations']['location'][0]['locationName'])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for location in locations:
    time = location['weatherElement'][0]['time'][0]['dataTime']
    value = location['weatherElement'][0]['time'][0]['elementValue']['value']
    print('地區:' + location['locationName'] + '、第一筆時間:' + time + '、溫度:' + value)
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
firstlocation = len(locations[0]['weatherElement'])
print(locations[0]['locationName'])
for number in range(firstlocation):
    times = location['weatherElement'][number]['time']
    for time in times:
        ltime = time['dataTime']
        lvalue = time['elementValue']['value']
        print('時間:' + ltime, '溫度: ' + lvalue)
```
